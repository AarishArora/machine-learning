```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Data points
data = np.array([
    [1, 2], [1.5, 1.8], [5, 8], [8, 8],
    [1, 0.6], [9, 11], [8, 2], [10, 2],
    [9, 3], [4, 1], [1, 4], [5, 6], [1.9, 3.5]
])

# Step 2: Initialize centroids (pick the first k points)
k = 3  # Number of clusters
print("Initial Centroids:")
centroids = np.copy(data[:k])  # Use first k points as centroids
print(centroids)

# Step 3: Run K-Means for a fixed number of iterations
for epoch in range(5):  # Number of iterations
    print(f"\nEpoch {epoch + 1}")
    # Step 3.1: Assign points to the nearest centroid
    clusters = []
    for point in data:
        distances = [np.linalg.norm(point - centroid) for centroid in centroids]
        cluster = np.argmin(distances)
        clusters.append(cluster)
    clusters = np.array(clusters)

    # Step 3.2: Update centroids based on cluster means
    for i in range(k):
        points_in_cluster = data[clusters == i]
        if len(points_in_cluster) > 0:  # Avoid division by zero
            centroids[i] = np.mean(points_in_cluster, axis=0)
    print("Updated Centroids:")
    print(centroids)

# Step 4: Plot the final clusters and centroids
colors = ['r', 'g', 'b', 'y', 'c', 'm']
for i in range(k):
    points_in_cluster = data[clusters == i]
    plt.scatter(points_in_cluster[:, 0], points_in_cluster[:, 1], color=colors[i % len(colors)], label=f'Cluster {i}')
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, marker='X', c='black', label='Centroids')
plt.legend()
plt.title('K-Means Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
